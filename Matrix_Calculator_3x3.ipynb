```python
# ============================================
# 3x3 MATRIX CALCULATOR PROJECT
# Developed using NumPy and Matplotlib
# This program performs all basic and advanced
# matrix operations with visualization
# ============================================

import numpy as np
import matplotlib.pyplot as plt
```

```python
# --------- TAKE INPUT FOR MATRIX A ----------
print("Enter values for 3x3 Matrix A")

A = []  # empty list to store rows of matrix A

for i in range(3):
    row = list(map(float, input(f"Enter row {i+1} (3 values separated by space): ").split()))
    A.append(row)

# convert list into NumPy array
A = np.array(A)

print("\nMatrix A is:")
print(A)
```

```python
# --------- TAKE INPUT FOR MATRIX B ----------
print("\nEnter values for 3x3 Matrix B")

B = []  # empty list to store rows of matrix B

for i in range(3):
    row = list(map(float, input(f"Enter row {i+1} (3 values separated by space): ").split()))
    B.append(row)

# convert list into NumPy array
B = np.array(B)

print("\nMatrix B is:")
print(B)
```

```python
### --------- TAKE SCALAR VALUE ----------
scalar = float(input("\nEnter a scalar value: "))
```

```python
# MATRIX CALCULATIONS


# Addition of matrices
addition = A + B

# Subtraction of matrices
subtraction = A - B

# Matrix multiplication
multiplication = A @ B

# Scalar multiplication
scalar_multiplication = scalar * A

# Determinant of matrix A
det_A = np.linalg.det(A)
```

```python
# Inverse of matrix A (only if determinant is not zero)
if det_A != 0:
    inverse_A = np.linalg.inv(A)
else:
    inverse_A = "Inverse not possible because determinant is zero"
```

```python
# Transpose of matrix A
transpose_A = A.T

# Rank of matrix A
rank_A = np.linalg.matrix_rank(A)
```

```python
# Trace of matrix A
trace_A = np.trace(A)

# Eigenvalues and Eigenvectors of matrix A
eigen_values, eigen_vectors = np.linalg.eig(A)
```

```python
# Solving linear equation AX = B
if det_A != 0:
    solution_X = np.linalg.solve(A, B)
else:
    solution_X = "No unique solution exists"
```

```python
# DISPLAYING ALL RESULTS

print("\n========== MATRIX CALCULATOR OUTPUT ==========")

print("\n1. Matrix Addition (A + B):")
print(addition)

print("\n2. Matrix Subtraction (A - B):")
print(subtraction)

print("\n3. Matrix Multiplication (A × B):")
print(multiplication)

print("\n4. Scalar Multiplication:")
print(scalar_multiplication)

print("\n5. Determinant of Matrix A:")
print(det_A)

print("\n6. Inverse of Matrix A:")
print(inverse_A)

print("\n7. Transpose of Matrix A:")
print(transpose_A)

print("\n8. Rank of Matrix A:")
print(rank_A)

print("\n9. Trace of Matrix A:")
print(trace_A)

print("\n10. Eigenvalues of Matrix A:")
print(eigen_values)

print("\n11. Eigenvectors of Matrix A:")
print(eigen_vectors)

print("\n12. Solution of AX = B:")
print(solution_X)
```

```python
# Heatmap representation of Matrix A
plt.imshow(A)
plt.colorbar()
plt.title("Heatmap Representation of Matrix A")
plt.show()
```

```python
# Bar graph of eigenvalues
plt.bar([1, 2, 3], eigen_values)
plt.xlabel("Eigenvalue Number")
plt.ylabel("Eigenvalue")
plt.title("Eigenvalues of Matrix A")
plt.show()
```

```python
# Eigenvector visualization (2D projection)
plt.quiver(0, 0, eigen_vectors[0, 0], eigen_vectors[1, 0], scale=1)
plt.quiver(0, 0, eigen_vectors[0, 1], eigen_vectors[1, 1], scale=1)
plt.quiver(0, 0, eigen_vectors[0, 2], eigen_vectors[1, 2], scale=1)

plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid()
plt.title("Eigenvector Direction Visualization")
plt.show()
```
